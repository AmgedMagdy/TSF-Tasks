# **Prediction using Supervised ML**
### **Simple Linear Regression**
In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

```python
# Importing required libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Importing data
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
s_data = pd.read_csv(url)
s_data.head(5)

```

```python
# Plotting Hours vs Scores
x = s_data['Hours'].to_list()
y = s_data['Scores'].to_list()

plt.scatter(x, y)
plt.xlabel('Hours')  
plt.ylabel('Scores')  
plt.show()
```

**From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.**

### **Training the Algorithm**


```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(x, y)

def calc_y(x):
  return slope * x + intercept

mymodel = list(map(calc_y, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.xlabel('Hours')  
plt.ylabel('Scores')  
plt.show()
```

### **Making Predictions**


```python
Scores_predicted = []
for hour in x:
    y_predicted = calc_y(hour)
    Scores_predicted.append(y_predicted)
    
df = pd.DataFrame({'Actual': y, 'Predicted': Scores_predicted})  
df.head(10)
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y, Scores_predicted)) 
```
